- checking for duplicate and null unique identifiers

```python
df = spark.sql("""
               SELECT COUNT(*) null_or_duplicated_customer_id_count FROM
               teleco_churn_datalakehouse.bronze.teleco_customer_churn 
               WHERE CustomerID IS NULL OR Count != 1; 
               """).display()

#  finding
#  All customers ID's are unique / No duplicate customer ID's
```
