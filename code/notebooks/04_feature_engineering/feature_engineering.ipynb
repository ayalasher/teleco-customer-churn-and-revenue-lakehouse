```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("featureEngineering").getOrCreate()
```

```python
df = spark.table("teleco_churn_datalakehouse.gold.fact_customer_subscription")
```

```python
df.display()
```

```python
from pyspark.sql.functions import when

df_features = df.withColumn(
    "churn_label",
    when(df.churn_flag == 1, 1).otherwise(0)
)
```

```python
feature_df = df_features.select(
    "customer_key",
    "Tenure_Months",
    "Monthly_Charges",
    "Total_Charges",
    "churn_label"
)
```

```python
feature_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("teleco_churn_datalakehouse.gold.customer_churn_features")
```
